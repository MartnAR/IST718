```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('Case3SalesProducts.csv')
```

```python
dfSingapore = df[df['Country']=='Singapore']
```

```python
dfSingapore.describe()
```

```python
sns.violinplot(x='SustainableClaim', y='GrossMargin', data=dfSingapore, hue='SustainableClaim')
_ = plt.legend(loc='center right', bbox_to_anchor=[1.3, 0.5])
plt.show()
```

```python
sns.violinplot(x='SustainableMarketing', y='GrossMargin', data=dfSingapore, hue='SustainableMarketing')
_ = plt.legend(loc='center right', bbox_to_anchor=[1.3, 0.5])
plt.show()
```

```python
sns.violinplot(x='SustainableClaim', y='GrossMargin', data = df, hue='SustainableClaim')
_ = plt.ylim([0.2, 1])
plt.show()
```

```python
sns.violinplot(x='SustainableMarketing', y='GrossMargin', data = df, hue='SustainableMarketing')
_ = plt.ylim([0, 1])
plt.show()
```

```python
logreg = LogisticRegression()

dfSingaporeNa = dfSingapore.dropna()

model = logreg.fit(dfSingaporeNa[['GrossMargin', 'Revenue']], dfSingaporeNa['SustainableClaim'])
print(model.coef_)

print(model.score(dfSingaporeNa[['GrossMargin', 'Revenue']], dfSingaporeNa['SustainableClaim']))
```
