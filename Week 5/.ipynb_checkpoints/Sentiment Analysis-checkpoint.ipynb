```python
# Packages to be used.
import os
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import re 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
import collections
```

```python
# Import the corpus that will be analyzed: Baseball Prospectus' 2019 Oakland Athletics Top 10 Prospects list. 
corpus = pd.read_csv('Corpus/Oakland Athletics 2019 Top 10 Prospects.csv')

# Just look at the first line of the data frame. 
corpus.iloc[0]
```

```python
# Looking at the data frame, we just want the prospect name and their report. 
corpus_simp = corpus[['prospect', 'report']]
print(corpus_simp)
```

```python
# We'll now remove special characters and lower-case all words in the report column. We'll then remove stop words, stem, and lemmatize
# the report column. 
clean_text = []

for i in corpus_simp.report: 
    clean = [re.sub('[\W\_]', ' ', w) for w in i.split('\n')]
    lower = [y.lower() for y in clean]
    clean_text.append(lower)

# Print the first text to verify that it has been cleaned. 
print(clean_text[0])
```

```python
# Initiate English stop words
stop = stopwords.words('english')

# New list to house tokenized sentences and removed stopwords. 
rmv_stop_words = []

for i in clean_text: 
    i = str(i)
    i = re.sub(r"[^\w]", ' ', i)
    word_tokens = word_tokenize(i)
    no_stop_words = ' '.join([x for x in word_tokens if x not in stop])
    rmv_stop_words.append(no_stop_words)
    
# Print first sentence to compare with previous complete sentence. 
print(rmv_stop_words[0])
```

```python
# List to append tokenized strings
tokens = []

# Tokenize stirngs with removed stop words
for i in rmv_stop_words:
    token = word_tokenize(i)
    tokens.append(token)

# Cancatenate with original data frame
token_df = pd.DataFrame(np.array(tokens).reshape(-1, 1))
corpus_df = pd.concat((corpus_simp, token_df), axis=1, ignore_index=True)
corpus_df.columns = ['name', 'report', 'token_report']
```

```python
# Using only the first report, that of Oakland pitching prospect Jesus Luzardo, we'll build a dictionary and create a word cloud. 
luzardo = corpus_df.iloc[0, 2]

wordcount = {}

# Iterate over the word corpus to remove special characters that may muddy the data. 
for word in luzardo:
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1

# Print the 20 most common words. 
n_print = 20            
word_counter = collections.Counter(wordcount)
for word, count in word_counter.most_common(n_print):
    print(word, ": %d" % count)

# Create a data frame of the most common words 
# Draw a bar chart
lst = word_counter.most_common(n_print)
df = pd.DataFrame(lst, columns = ['Word', 'Count'])
df.plot.bar(x='Word',y='Count')
```

Obviously, the most common word has to be Luzardo. But we also see plus, one, breaking, and ball among the most common words. Perhaps the prospect writers are describing Luzardo's plus breaking ball that may be his ticket to the big leagues. 
We'll do one more graph, looking at the word cloud for Luzardo.

```python
for a, x in df.values:
    d[a] = x

# Adjust image parameters to increase height and width of figure, create word cloud, and show image
plt.rcParams['figure.figsize'] = [10, 5]
wordcloud = WordCloud(scale=5)
wordcloud.generate_from_frequencies(frequencies=d)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
